```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Operator
```

```python
qc=QuantumCircuit(2)
qc.cx(1,0)
qc.draw('mpl')
```

```python
unitary=Operator(qc)
unitary.draw('latex')
```

```python
qc=QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.cx(0,1)
qc.h(0)
qc.h(1)
qc.draw('mpl')
unitary=Operator(qc)
unitary.draw('latex')
```
