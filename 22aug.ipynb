```python
from qiskit import QuantumCircuit
import numpy as np
```

```python
qc= QuantumCircuit(3)
qc.x(0)
qc.h(1)
qc.z(2)
qc.ry(np.pi/4,0)
qc.crx(np.pi/4,0,1)
qc.mcry(-np.pi/8,[0,1],2)
qc.draw('mpl')
```

```python
print("width=",qc.width())
print("Depth=",qc.depth())
```

```python
from qiskit.quantum_info import Statevector
qc=QuantumCircuit(2)
qc.x(1)
qc.x(0)
qc.draw("mpl")
```

```python
state=Statevector.from_instruction(qc)
state.draw('latex')
```

```python
qc=QuantumCircuit(4)
qc.x(0)
qc.z(1)
qc.cx(1,2)
qc.rx(np.pi/12,2)
qc.crx(-np.pi/7,0,3)
qc.draw('mpl')
```

```python
state=Statevector(qc)
state
```

```python
state.draw('latex')
```

```python
qc=QuantumCircuit(4)
qc.x(2)
qc.ry(np.pi/8,3)
qc.draw('mpl')
```

```python
state2=state.evolve(qc)
state.draw('latex')

```

```python
from qiskit.visualization import plot_state_city
plot_state_city(state)
```
